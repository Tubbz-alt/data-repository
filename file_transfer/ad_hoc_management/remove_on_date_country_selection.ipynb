Info about [files to delete](https://github.com/enram/data-repository/issues/46):

```python
countries = ["be", "ch", "cz", "de", "dk", "ee", "es", "fr", "hr", "no", "pl", "si", "sk"]
start_date = datetime.date(2016, 11, 1)
end_date = datetime.date(2017, 11, 21)
```

Loop h5 files and remove those according to criteria:

```python
files_to_delete = []
```

```python
for j, file in enumerate(s3handle.bucket.objects.all()):
    if file.key.endswith(".h5"):
        file_info = parse_filename(os.path.basename(file.key))
        date = datetime.date(int(file_info["year"]), 
                             int(file_info["month"]), 
                             int(file_info["day"]))
        country = file_info["country"]
        if (country in countries) and (date > start_date) and (date <= end_date):
            files_to_delete.append(file)
            file.delete()
```

```python
deleted_files = [file_info.key for file_info in files_to_delete]
deleted_files
```
